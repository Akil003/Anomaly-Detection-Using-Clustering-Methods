# Isolation Forest

## Import Pre-processed data

```python
import pandas as pd

data = pd.read_csv("data/processed_data.csv")
data.shape
```

## Feed into Model

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(n_estimators=100, contamination=0.02, random_state=42)
history = model.fit(data)

# Predicting the anomalies
predictions = model.predict(data)

# Adding the predictions to the original data
data['anomaly'] = predictions
```

```python
len(data.loc[data['anomaly'] == -1]), len(data.loc[data['anomaly'] == 1])
```

```python
data.columns = ['x', 'y', 'label']
data.columns
```

```python
data['label'].unique()
```

## Visualize

```python
import matplotlib.pyplot as plt

# Get the unique cluster labels
labels = data['label'].unique()


plt.scatter(data.loc[data['label'] == -1]['x'], data.loc[data['label'] == -1]['y'], marker='.', color='black', s = 2)
plt.scatter(data.loc[data['label'] == 1]['x'], data.loc[data['label'] == 1]['y'], marker='o', color='cyan', s = 2)

# Set the plot title and axis labels
plt.title('Isolation Forest Result')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

## Result

Hence, 1000 network packets have been identifed to be anomalous and 48990 have been identified as normal. The blue node indicate normal traffic and black ones indicate abnormal ones.
